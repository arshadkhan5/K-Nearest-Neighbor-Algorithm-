#**K Nearest Neighbour for Fruit Detection**

```python
#Required Libraries
import numpy as np  
import pandas as pd 
import matplotlib.pyplot as plt 

```

```python
#Loading the data from GOOGLE DRIVE 
dataset = pd.read_csv("/content/drive/MyDrive/fruit_dataset.csv")
```

```python
#Showing  data
dataset
```

Here see that apple is labbeled as 1 ..mandarin as 2..Orange is 3 and lemon is 4

```python
#Showing the information about data 

dataset.info()
```

```python
#Spliting the data into X and Y 
X = dataset[['mass','width','height','color_score']]
Y = dataset['fruit_label']
```

```python
X
```

```python
Y
```

```python
#Spliting the data in train and test 
from sklearn.model_selection import train_test_split

x_train ,x_test , y_train ,  y_test = train_test_split(X ,Y , test_size=.10 , random_state= 0) # test_size for separate test data 

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#importing the K Nearest Neighbour
from sklearn.neighbors import KNeighborsClassifier

#Intializing the KNN
knn = KNeighborsClassifier(n_neighbors = 4 , metric = "minkowski") # minkowski is used to find encludian distance 

# Fitting the model 

knn.fit(x_train , y_train)
```

```python
#Prediction 
predict = knn.predict(x_test)

```

```python
print(predict)
```

###Prediction with unseen data
